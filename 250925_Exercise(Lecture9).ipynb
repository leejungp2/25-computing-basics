# Linear Search : find the index

```python
# make a random array
import random
listSize = 1000
nums = list(range(listSize))
random.shuffle(nums)
print(nums)
```

```python
# complete linear search function # 강의 자료 3가지 방식
def find_idx(L:list, num:int):
    for i in range(len(L)):
        if L[i] == num:
            return i
    return -1
```

```python
import time
total_time_list = []
for val in nums:
    start = time.time()*1000 # 코드 돌리기 전 시간
    idx = find_idx(nums, val)
    end = time.time()*1000 # 코드 돌리고 난 후
    total_time = end-start
    total_time_list.append(total_time)

sum_time = sum(total_time_list)
length = len(total_time_list)

print("Latency:", end-start, "ms")
```
