# Exercise 1

Given an integer greater than or equal to 0 as input, implement a function that returns the number of '1's when expressed in bits. 

Input integer does not exceed 2^31 - 1 and there should not be any leading zeros in the answer except for the zero itself.

```python
# 비트 연산 사용: 교수님 풀이
def numOfOnes(n: int) -> int:
    count = 0 # 비트 개수 카운터?
    while n: 
        count += n & 1   # 마지막 비트가 1인지 확인
        n >>= 1          # 오른쪽으로 한 칸 이동
    return count
```

```python
def binary(input_num:int)->int:
    
    # bin 못 떠올렸을 떄 안쓰고 하는 방법은 없나?
    
    n = bin(input_num)
    return bin(n).count('1')
```

```python
# 내가 처음 한 발상: 2로 나눠서 이진수로 만드는? %로 안풀어도 간소화 할 수 있는 이유 찾아보기
def count_ones_div(n):
    count = 0
    while n > 0:
        if n % 2 == 1:
            count += 1
        n //= 2
    return count
```

```python
# Test
print(numOfOnes(16))
print(numOfOnes(7))
print(numOfOnes(1234567))
```

```python
# Input: start = 10, goal = 7
# Output: 3
# Explanation: The binary representation of 10 and 7 are 1010 and 0111 respectively. We can convert 10 to 7 in 3 steps:
# - Flip the first bit from the right: 1010 -> 1011.
# - Flip the third bit from the right: 1011 -> 1111.
# - Flip the fourth bit from the right: 1111 -> 0111.
# It can be shown we cannot convert 10 to 7 in less than 3 steps. Hence, we return 3.
```

```python
def minBitFlips(start: int, goal: int) -> int:
    # 로직
    # idea 1: 다른 걸 1로 표현한 후, 1의 개수 세기 (문제 1과 마찬가지로 하나씩 옮겨가면서 확인))
    # idea 2: xor: 하나가 0, 하나가 1일 때만 1
    return numOfOnes(start^goal)
```
