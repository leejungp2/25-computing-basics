```python
# 풀이 3
def findprefix(number:list) -> bool:
    hash = {}
    
    # 1. 모든 번호를 해시 테이블에 저장
    for num in number:
        hash[num] = True
    
    # 2. 각 번호 접두사 존재하는지 검사
    for num in number:
        prefix = "" # prefix: num의 앞부분을 하나씩 늘려가며 만든 부분 문자열
        for ch in num[:-1]: # 마지막 숫자 제외
            prefix += ch
            if prefix in hash:
                return False
    return True
```

```python
# 풀이 4: sort 사용
def findprefix(number:list) -> bool:
    # 1. 정렬
    number.sort()

    # 2. 인접한 두 문자열만 비교
    for i in range(len(number) - 1):
        # 다음 문자열이 현재 문자열로 시작하면 prefix 관계 존재
        if number[i+1].startswith(number[i]):
            return False

    return True
```
